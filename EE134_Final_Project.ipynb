```python
import matplotlib.pyplot as plt
import numpy as np 
import csv
```

```python
bt = []
voc = []
with open('/Users/nikitamehendale/Downloads/EE 134 Project Data - Thickness.csv') as csvfile:
    csv_data = list(csv.reader(csvfile))
    for k in range(1, 18):
        bt.append(float(csv_data[k][0]))
        voc.append(float(csv_data[k][1]))
plt.scatter(bt, voc)
plt.title('Voc vs Base Thickness')
plt.xlabel('Base Thickness (cm)')
plt.ylabel('Voc (mV)')
#highilight maximum value on plot
max_voc = max(voc)
max_voc_index = voc.index(max_voc)
plt.scatter(bt[max_voc_index], voc[max_voc_index], color='red')
#include legend saying red dot is max value
plt.legend(['VoC', 'Max Voc'])
```

```python
bt = []
jsc = []
with open('/Users/nikitamehendale/Downloads/EE 134 Project Data - Thickness.csv') as csvfile:
    csv_data = list(csv.reader(csvfile))
    for k in range(1, 18):
        bt.append(float(csv_data[k][0]))
        jsc.append(float(csv_data[k][2]))
plt.scatter(bt, jsc)
plt.title('Jsc vs Base Thickness')
plt.xlabel('Base Thickness (cm)')
plt.ylabel('Jsc (mA/cm^2)')
#highilight maximum value on plot, if multiple max values higlight all of them
max_jsc = max(jsc)
max_jsc_index = []
for i in range(len(jsc)):
    if jsc[i] == max_jsc:
        max_jsc_index.append(i)
for i in max_jsc_index:
    plt.scatter(bt[i], jsc[i], color='red')
plt.legend(['Jsc', 'Max Jsc'])
plt.show()
```
